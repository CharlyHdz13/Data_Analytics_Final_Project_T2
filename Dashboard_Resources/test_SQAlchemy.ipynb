```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
comorbilities = Base.classes.comorbilities
```

```python
# Create a session
session = Session(engine)
```

```python
test=session.query(comorbilities.smoking).all()
```

```python
# Use SQL to query 'all' the records in the the Database 
data = engine.execute("SELECT * FROM comorbilities")
for row in data:
    print(row)
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)
# Using the inspector to print the column names within the 'dow' table and its types
columns = [column['name'] for column in inspector.get_columns('comorbilities')]
```

```python
columns
```

```python
#  Create a meta data object to hold the reflected table schema
metadata = MetaData()
```

```python
# Create a table object and use 'autoload' and 'autoload_with' to define the columns from the table. 
table = Table('comorbilities', metadata, autoload=True, autoload_with=engine)
```
